```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its derivative
def f(x):
    return x**2 + 4*x + 5

def f_derivative(x):
    return 2*x + 4

# Gradient Descent Parameters
learning_rate = 0.1
iterations = 50
x_current = 10  # Initial guess
history = []    # To store x values for visualization

# Gradient Descent Algorithm
for i in range(iterations):
    gradient = f_derivative(x_current)
    x_current -= learning_rate * gradient
    history.append(x_current)

# Minimized value and location
min_x = x_current
min_value = f(min_x)

print(f"Global Minima at x = {min_x}, f(x) = {min_value}")

# Plotting
x = np.linspace(-10, 10, 100)
plt.plot(x, f(x), label="Function f(x)")
plt.scatter(history, [f(i) for i in history], color='red', label="Gradient Descent Steps")
plt.title("Gradient Descent for f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
# Simple Linear Regression

class MeraLR:
    
    def __init__(self):
        self.m = None
        self.b = None
    
    def fit(self,X_train,y_train):
        num = 0
        den = 0
        
        #  X : features ( indepdenet valriaes ) y : targets (depdent variables) 
        
        for i in range (X_train.shape[0]):    
           num = num + ((X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean()))
           den = den + ((X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean()))
            
        self.m = num/den
        self.b = y_train.mean() - (self.m * X_train.mean())
        
        # calculate slope m and intercept b 
        
        print(self.m)
        
        print(self.b)
        
    def predict(self, X_test):
        #  y = m .x + b
        return self.m * X_test + self.b
```
